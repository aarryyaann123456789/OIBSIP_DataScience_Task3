```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Load the data
data = pd.read_csv('car data.csv')

# 2. CLEANING: Convert text to numbers
# We use the exact column name 'Selling_type' here
data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}}, inplace=True)
data.replace({'Selling_type':{'Dealer':0,'Individual':1}}, inplace=True)
data.replace({'Transmission':{'Manual':0,'Automatic':1}}, inplace=True)

# 3. Define X and Y
# X = Features (Everything except the price and name)
# Y = Target (Selling_Price)
X = data.drop(['Car_Name','Selling_Price'], axis=1)
Y = data['Selling_Price']

# 4. Split and Train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, Y_train)

# 5. Visualize Results
plt.scatter(Y_train, lin_reg_model.predict(X_train))
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()

print("SUCCESS! The model is trained and the graph should be visible above.")
```
